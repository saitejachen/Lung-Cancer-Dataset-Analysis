```python
import numpy as np
import pandas as pd
from scipy.io.arff import loadarff 

raw_data = loadarff('C:/Users/Sai/Desktop/Jupyter Notebooks/datasets/Lung Cancer/Lung.arff')
df_data = pd.DataFrame(raw_data[0])
```

```python
df_data['type']=df_data['type'].astype(int)
```

```python
df_data.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
data=df_data.drop(['type'],axis=1)
```

```python
target=pd.DataFrame(df_data['type'],index=df_data.index,columns=['type'])
target
```

```python
data_scaled=scaler.fit_transform(data)
```

```python
data_scaled=pd.DataFrame(data_scaled,index=data.index,columns=data.columns)
```

```python
data_scaled.head()
```

```python
X=data_scaled
```

```python
X['type']=target
```

```python
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
```

```python
X.head()
```

```python
pca=PCA()
pca.fit(X)
cumsum=np.cumsum(pca.explained_variance_ratio_)
d=np.argmax(cumsum>=0.95)
print(d)
pca=PCA(n_components=d)
X_reduced=pca.fit_transform(X)
```

```python
X_reduced=pd.DataFrame(X_reduced,index=X.index,columns=range(d))
```

```python
X_reduced.head()
```

```python
y=target
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_reduced,y,test_size=.2,random_state=5)
```

```python
from sklearn.svm import SVC
svm_clf=SVC(gamma=.1,class_weight ='balanced',random_state=25)
y_train=np.array(y_train).ravel()
svm_clf.fit(X_train,y_train)
```

```python
from sklearn.model_selection import cross_val_predict,cross_val_score
from sklearn.metrics import confusion_matrix
import seaborn as sn

print(svm_clf.score(X_test,y_test))
y_pred=cross_val_predict(svm_clf,X_reduced,y,cv=3)
print(cross_val_score(svm_clf,X_reduced,y,cv=3))
cm=confusion_matrix(y,y_pred)
sn.heatmap(cm,annot=True,fmt='d')
```

```python
from sklearn.model_selection import GridSearchCV
parameters={'kernel':('linear','rbf','poly'),'C':[0.01,0.1,1,10],'gamma':[0.01,0.1,1,10]}
grid_search_svm=GridSearchCV(SVC(class_weight='balanced'),parameters)
grid_search_svm.fit(X_train,y_train)
print(grid_search_svm.score(X_test,y_test))
grid_search_svm.best_estimator_
y_pred=cross_val_predict(grid_search_svm,X_reduced,y,cv=3)
print(cross_val_score(grid_search_svm,X_reduced,y,cv=3))
cm=confusion_matrix(y,y_pred)
sn.heatmap(cm,annot=True,fmt='d')
```

```python
kernel=['linear','rbf','poly']
C=[0.01,0.1,1,10]
gamma=[0.01,0.1,1,10]
```

```python
import matplotlib.pyplot as plt
color=['red','blue','green']
size=[100,200,300]
```

```python
s=[]
for k in kernel:
    for c in C:
        for g in gamma:
            gs_clf=SVC(kernel=k,C=c,gamma=g)
            gs_clf.fit(X_train,y_train)
            s.append(gs_clf.score(X_test,y_test))      
```

```python
plt.plot(s) # linear performs the best
```

```python
from sklearn.linear_model import LogisticRegression
lg_clf=LogisticRegression(solver='lbfgs',multi_class='multinomial',class_weight='balanced')
lg_clf.fit(X_train,y_train)
lg_clf.score(X_test,y_test)
y_pred=cross_val_predict(lg_clf,X_reduced,y,cv=3)
print(lg_clf.score(X_test,y_test))
print(cross_val_score(lg_clf,X_reduced,y,cv=3))
cm=confusion_matrix(y,y_pred)
sn.heatmap(cm,annot=True,fmt='d')
```
